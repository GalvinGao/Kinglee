Let each mass be $ 1 \textrm{kg}$ and each spring have spring constant $ 100 \textrm{N/m}$ with rest length $1 \textrm{m}$. Let the distance between the walls be $l$. $ F $ is an applied force on $m_1$. Let the friction Coefficent be $C=3$.

$$ \sum F_1 = F-k x_1 - k \left(x_1-x_2\right) - C v_1= m_1 \ddot{x_1}$$
$$ \sum F_2 = - k \left(x_2-x_1\right) - k \left( x_2 - l \right) - C v_2 = m_2 \ddot{x_2} $$


First let's just pull the blocks and watch them come to rest. This is done by changing the first element in the list for position.

```python
# importing the tools we need to use
import numpy as np
import matplotlib.pyplot as plt

# making a time step
steps=2000
C=3.
dT=0.001 
k=100.0
m1=1.
m2=1.
l=3

# function of force
def F(time):
    omega=15
    A=0.
    phi=1.
    return A*np.sin(omega*time+phi)

# making empty lists to hold the sequences
t=[0]*steps
x1=[0]*steps
v1=[0]*steps
a1=[0]*steps
x2=[0]*steps
v2=[0]*steps
a2=[0]*steps

# putting in our initial values
x1[0]=0.8
v1[0]=0.
x2[0]=2.2
v2[0]=0.

# using the linear equations and first equation to fill our lists
for n in range(steps-1):
    a1[n]=(F(n*dT)-k*x1[n]-k*(x1[n]-x2[n])-C*v1[n])/m1
    a2[n]=(-k*(x2[n]-x1[n])-k*(x2[n]-l)-C*v2[n])/m2
    t[n+1]=dT*(n+1)
    x1[n+1]=x1[n]+v1[n]*dT
    v1[n+1]=v1[n]+a1[n]*dT
    x2[n+1]=x2[n]+v2[n]*dT
    v2[n+1]=v2[n]+a2[n]*dT
    
# position graph
plt.plot(t,x1,t,x2)
plt.show()
```
